### Category and Sub-Category
#### Sales and Profit

```python
x= data.groupby(["Category","Sub-Category"]).Sales.sum()
y= data.groupby(["Category","Sub-Category"]).Profit.sum()
```

### Sales

```python
fig = plt.figure()
axe0=fig.add_subplot(2,2,1)
axe1=fig.add_subplot(2,2,2)
axe2=fig.add_subplot(2,2,3)
axe3=fig.add_subplot(2,2,4)

#All sub_categories vs Sales
x.sort_values(ascending= True).plot(kind ="barh",figsize=(20,20),ax=axe0) #barh = Horizontal Bar
axe0.set_title("All Sub-Categories vs Sales")
axe0.set_ylabel("Sub-Categories")
axe0.set_xlabel("Sales")

#Furniture vs Sales
x["Furniture"].sort_values(ascending = True).plot(kind = "barh",figsize = (20,20),ax=axe1)
axe1.set_title("Furniture vs Sales")
axe1.set_ylabel("Furniture")
axe1.set_xlabel("Sales")

#Technology vs Sales
x["Technology"].sort_values(ascending = True).plot(kind = "barh",figsize =(20,20),ax= axe2)
axe2.set_title("Technology vs Sales")
axe2.set_ylabel("Technology")
axe2.set_xlabel("Sales")

# Offlice Supplies vs Sales
x["Office Supplies"].sort_values(ascending = True).plot(kind = "barh",figsize=(20,20),ax= axe3)
axe3.set_title("Office Supplies vs Sales")
axe3.set_xlabel("Sales")
axe3.set_ylabel("Office Supplies")
```

## Analysis and Recommendations for Business Improvement

**1. Relationship between Discount, Sales, and Profit:**
   - It is observed that an increase in discounts leads to a decrease in both profit and sales.
   - Additionally, an increase in sales results in higher profits.

**2. Correlation between Key Metrics:**
   - The most significant correlation exists between Sales and Profit, suggesting a strong positive relationship.
   - The least correlation is found between Discount and Quantity, indicating a weak or negligible association.

**3. Ship Mode Preferences:**
   - The majority of customers prefer the "Standard Class" ship mode.
   - The least preferred ship mode is "Same Day."

**4. Profit and Sales by Ship Mode:**
   - The "Standard Class" ship mode generates the highest profit and sales.
   - Conversely, the "Same Day" ship mode yields the lowest profit and sales.

**5. Business Distribution by Category:**
   - The primary focus of the business lies in the "Office Supplies" category.
   - The "Technology" category receives the least attention.

**6. Profit and Sales by Category:**
   - The "Technology" category records the highest profit and sales.
   - In contrast, the "Office Supplies" category reports the lowest profit and sales.

**7. Business Distribution by Sub-Category:**
   - The majority of business activities are centered around "Binders," "Paper," and "Furnishings."
   - "Copiers," "Machines," and "Supplies" have the least business involvement.

**8. Sales and Profit by Sub-Category:**
   - "Copiers," "Phones," and "Accessories" within the "Technology" category generate the highest sales and profit.
   - The "Tables," "Bookcases" in the "Furniture" category, and "Supplies" in the "Office Supplies" category contribute the least to sales and profit.

**9. Geographic Distribution of Buyers:**
   - The states with the most buyers are "California," "New York," and "Texas."
   - The states with the fewest buyers are "Wyoming," "West Virginia," and "North Dakota."

**10. Sales and Profit by City:**
    - The cities of "California" and "New York" report the highest sales and profit.
    - Conversely, cities such as "Oregon," "Florida," "Arizona," "Tennessee," "Colorado," "North Carolina," "Illinois," "Pennsylvania," "Ohio," and "Texas" face lower sales and profit margins.

**11. Business Distribution by Customer Segment:**
    - The majority of business occurs in the "Consumer" segment.
    - The "Home Office" segment contributes the least to business operations.
    
**12. Profit and Sales by Customer Segment:**
    - The "Consumer" segment leads in both profit and sales, while the "Home Office" segment lags behind.

**13. Regional Business Distribution:**
    - The business predominantly operates in the Western Region and minimally in the Southern Region.
    - The Western and Eastern Regions contribute the most to profit, while the Central and Southern Regions generate lower profits.

**14. Impact of Discounts on Profit:**
    - Discounts and profit share a negative relationship, indicating that increased discounts lead to reduced profits.
    
**15. Impact of Discounts on Sales:**
    - Although the relationship between discounts and sales is weaker, increasing discounts still leads to decreased sales.

**Conclusion and Recommendations for Improvement:**
   - To enhance profitability and sales, consider reducing discounts, especially on non-essential items.
   - Focus on the "Standard Class" ship mode, which has proven to be the most profitable.
   - Shift the business's emphasis towards the "Technology" category, particularly "Copiers," "Phones," and "Accessories."
   - Implement strategies to improve sales in cities with lower performance, such as "Oregon," "Florida," "Arizona," and others.
   - Continue to prioritize cities with the highest buyers, sales, and profit, such as "California" and "New York."
   - Encourage and target the "Consumer" segment, which is the most profitable.
   - Invest efforts in expanding business in the Western and Eastern Regions to maximize profit potential.

This comprehensive analysis and set of recommendations can serve as a valuable reference for business improvement efforts, providing a clear path forward for enhanced profitability and success.##
